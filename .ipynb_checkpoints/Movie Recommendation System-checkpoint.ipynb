What is Movie Recommendation System?
A movie recommendation system, uses machine learning to predict or filter user's film preferences based on their prior decisions and actions. 

➡️ How does the movie recommendation system work?
Every recommender system primarily consists of two components: users and items. Users receive movie predictions from the system, and the actual movies are the products.
Filtering and predicting only the movies that a matching user is most likely to wish to see is the main objective of movie recommendation systems. The user information from the system's database is used by the ML algorithms for these recommendation systems. Based on information from the past, this data is used to forecast the user  behavior in the future.

➡️ Filtering Strategies of recommendation system
- Filtering based on content - 
A method of filtering movies in movie recommendation systems that makes advantage of the items' data (movies). This information, which is taken from just one user, is quite important in this case. This technique uses an ML algorithm to suggest movies that are comparable to the user's past choices. Therefore, the information about the prior movie choices and likes of just one person is used to generate similarity in content-based filtering.


- Collaboration Filtering -
As the name implies, this filtering technique is based on the interactions between the relevant person and other users. For the best outcomes, the system contrasts and compares these behaviors. It combines the film choices and usage patterns of several people.

Collaboration Filtering has two types

1.Collaboration Filtering based on users
goal is to find pattern in target users and other database users like movie preference

2. Collaboration Filtering based on items (items means movies)
find comparabale products(movies) that target users rate or interacts with


* This recommandation system uses **Filtering based on content** - it uses features such as cast , creaw , keywords ,overview to find similarities between movies. .
<br>
<br>
* **Unsupervised Machine Learning Algorithm** - This recommandation system uses usupervised system. In this case, the algorithm computes similarities between movies based on their features (possibly descriptions, genres, etc.) and recommends similar movies to a given input movie. There's no explicit supervision or labeled output guiding the learning process.

```python
# abstarct syntax tree (ast)
import ast
```

The 'ast.literal_eval' function in Python safely evaluates a string containing a Python literal (like a number, list, or dictionary) into its corresponding Python object.

```python
# convert littral to object - converting string representation of list of dictioneries in genre column into list of dictioneries
# then append value of a key'name' into empty list 'l'
# conver string to list or dictioneries etc whatever is there inside a string
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
        
```

```python
# use '.apply(funtion_name)' function to apply funtion on raw of dataset.

movies_df['genres']=movies_df['genres'].apply(convert)
movies_df['keywords']=movies_df['keywords'].apply(convert)
```

```python
movies_df.head()
```

```python
# extracting only 3 cast names
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter = counter + 1
        else:
            break
    return L
        
```

```python
movies_df['cast']=movies_df['cast'].apply(convert3)
```

```python
movies_df.head()
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
    return L
        
```

```python
movies_df['crew']=movies_df['crew'].apply(fetch_director)
```

```python
# feature data extraction using count factorizer  - extracting important features or variable from data (from tag column)
# we are extracting count of most frequencty occuring words from tag column. those words are features of our model.

from sklearn.feature_extraction.text import CountVectorizer


# used to transfrom given text into matrix on basis of frequency count of each word occues in entire text
```

```python
cv = CountVectorizer(max_features = 5000,stop_words='english')

# max_features = 5000 - 5000 most frequntly occuring words would be saved in matrix
# stop_words = ignores stop_words from english language like is , the , and
```

```python
# Applying logic stored in cv varibale(in above cell) to 'tags' column of new_df
# .toarray() - convertes the reuslting sparse matrix into numpy array

cv.fit_transform(new_df['tags']).toarray().shape

# rows - 4805  -> total rows of tag column
# column 5000 -> for 5000 unique words and there count  in each row of tag column
```

* a tabel(matrix)will be created. 
* column headings - 5000 most frequently occuring unique words from tags column
* row heading - each sentence from tag column. eg, row one = first row of tag column

```python
# learns the vocabulary (fit), and then transforms the text data into a matrix of token counts.

vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
import nltk
# natural language tool kit
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
# process of reducing a word to its word org. form. eg - 'running' becomes run
```

```python
def stem(text):   # Define the stem function
    y = []        # Initialize an empty list to store stemmed words
    for i in text.split():   # split the text into comma seperated list , Loop through each word in the input text
        y.append(ps.stem(i))    # Apply stemming to the current word and append to the list
    return " ".join(y)         # Join the stemmed words back into a string again and return
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
# new_df['tags'][1]
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# co sign similarity measures similarity between two factors 
# used to meausre document similarity in text analysis
```

* Here we are measuring similarites in tag column means how similar each one sentence is to all other sentences of tag column.
* sentences means each row of tag column. each row has one sentence.
* this funtion will return array. on row heading and clumn heading we have sentences of tag column
* inside matric we have valuees as per thier similarity (see other example in learning file)

```python
cosine_similarity(vectors)    # 1 means completely similear(same sentence)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity[0].shape
```

 sorting a list, list which has index and value of similarity[0] , sorting is revrse order , key starts from 1 index , 
 after reversing it returns top 5 elememnts of the list through slicing -  [1:6]

* enumerate - iterate over a list or tuple and creates tuple with index
* converted this tuple into list
* sorted this list into reverse order
* retrive only top 5 values

```python

sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    try:
        movie_index = new_df[new_df['title']==movie].index[0]  # selects first raw that matches the input movie. returns raw index
        distances = similarity[movie_index]
        movies_list = sorted(list(enumerate(distances)),reverse = True , key = lambda x:x[1])[1:6]
     
        for i in movies_list:
            print(new_df.iloc[i[0]].title)
    except IndexError:
        print(f"Movie '{movie}' not found in the dataset. Please enter a valid movie title.")
       
        
```

```python
from ipywidgets import interact, widgets
import voila


def recommend(movie):
    movie_indices = new_df[new_df['title'] == movie].index
    if len(movie_indices) == 0:
        print(f"Movie '{movie}' not found in the dataset. Please enter a valid movie title.")
        return
    else:
        movie_index = movie_indices[0]
        distances = similarity[movie_index]
        movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

        print(f"Top 5 recommended movies for '{movie}':")
        for i in movies_list:
            print(new_df.iloc[i[0]].title)

# Create a dropdown with movie titles
movie_dropdown = widgets.Dropdown(
    options=new_df['title'].tolist(),
    description='Select a Movie:'
)

# Use interact to call the recommend function with the selected movie
interact(recommend, movie=movie_dropdown);

# To use Voila, save your notebook, and then run the following command in your terminal:
# voila your_notebook.ipynb
```

In summary, this code is using a collaborative filtering approach where recommendations are based on the similarity between movies. The similarity is computed using some pre-existing similarity matrix, and the top 5 similar movies are recommended for the input movie.
